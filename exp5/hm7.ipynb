```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to perform quadratic least-squares fit
def least_squares_quadratic_fit(x, y):
    m = len(x) - 1
    A = np.zeros((m + 1, 3))
    b = np.zeros((m + 1, 1))

    # Populate the coefficient matrix A and vector b
    for i in range(m + 1):
        A[i][0] = x[i] ** 2  # Quadratic term
        A[i][1] = x[i]       # Linear term
        A[i][2] = 1          # Constant term
        b[i][0] = y[i]       # Corresponding y value

    # Solve the system of equations using the least squares method
    coefficients = np.linalg.lstsq(A, b, rcond=None)[0]
    a, b, c = coefficients.flatten()  # Extract the quadratic, linear, and constant coefficients

    return a, b, c

# Given data
x_data = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
y_data = [0.2124, 1.2448, 1.3962, 2.4794, 2.6790, 4.4033, 4.8407, 3.6054, 5.4757, 5.0382, 5.8457]


# x_data = [-1, 0, 1, 2]
# y_data = [-3, 1, 6, 8]

# Call the function to compute the quadratic fit
a, b, c = least_squares_quadratic_fit(x_data, y_data)

# Generate points for the quadratic curve
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = a * x_fit ** 2 + b * x_fit + c

# Create a plot of the points and the quadratic function
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, label='Data Points', color='blue')
plt.plot(x_fit, y_fit, label=f'Quadratic Fit: y = {a:.4f}x^2 + {b:.4f}x + {c:.4f}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Quadratic Least-Squares Fit')
plt.show()
```
