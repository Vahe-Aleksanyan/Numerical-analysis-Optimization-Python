```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return 4 * x - x**2

def f_prime(x):
    return 4 - 2 * x

# Initial guess
x0 = 6

# Set the number of iterations
num_iterations = 5  # You can increase this number for more iterations.

# Lists to store the x and y values during iterations
x_values = [x0]
y_values = [f(x0)]

# Perform Newton's method iterations
for _ in range(num_iterations):
    x0 = x0 - f(x0) / f_prime(x0)
    x_values.append(x0)
    y_values.append(f(x0))

# Generate a range of x-values for plotting
x_range = np.linspace(min(x_values) - 1, max(x_values) + 1, 400)
y_range = f(x_range)

# Plot the function and Newton's iterations
plt.plot(x_range, y_range, label='f(x) = 4x - x^2')
plt.scatter(x_values, y_values, color='red', marker='o', label='Newton\'s iterations')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Newton\'s Method for f(x) = 4x - x^2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 4 * x - x**2

# Initial guesses
x0 = 3
x1 = 5

# Set the number of iterations
num_iterations = 5  # You can increase this number for more iterations.

# Lists to store the x and y values during iterations
x_values = [x0, x1]
y_values = [f(x0), f(x1)]

# Perform secant method iterations
for _ in range(num_iterations):
    x_new = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
    x_values.append(x_new)
    y_values.append(f(x_new))
    x0, x1 = x1, x_new

# Generate a range of x-values for plotting
x_range = np.linspace(min(x_values) - 1, max(x_values) + 1, 400)
y_range = f(x_range)

# Plot the function and the secant method iterations
plt.plot(x_range, y_range, label='f(x) = 4x - x^2')
plt.scatter(x_values, y_values, color='red', marker='o', label='Secant method iterations')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Secant Method for f(x) = 4x - x^2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# problem 4
def babylonian_sqrt(a):
    x_0 = a  # Initial guess
    x_values = [x_0]

    while True:
        x_next = 0.5 * (x_0 + a / x_0)  # Babylonian iteration formula
        x_values.append(x_next)

        if abs(x_next - x_0) < 1e-10:
            break

        x_0 = x_next  # Update x_0 after iteration

    return x_values

# Test
arr = [2, 4, 10]

for a in arr:
    x_values = babylonian_sqrt(a)
    print(f"Square root of {a} using Babylonian method:")
    for i, x in enumerate(x_values):
        print(f"Iteration {i}: x_{i} = {x:.10f}")
    print("\n")
```
